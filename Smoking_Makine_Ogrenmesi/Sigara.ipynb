```python
ilac = pd.read_csv("smoking.csv")
df = ilac.copy()
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
df["Cinsiyet"] = lbe.fit_transform(df["Cinsiyet"])
df["tartar"] = lbe.fit_transform(df["tartar"])


```

```python
df.corr()
```

```python
df.drop("sozlu", axis = 1, inplace=True)
```

```python
df.head()
```

```python
y = df["icmek"]
X = df.drop(['icmek'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
knn = KNeighborsClassifier()
knn_model = knn.fit(X_train, y_train)
knn_model
```

```python
y_pred = knn_model.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
cross_val_score(knn_model, X_test, y_test, cv = 10).mean()
```

Yapay Sinir Ağları

```python
from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.neural_network import MLPClassifier
mlpc = MLPClassifier().fit(X_train_scaled, y_train)
```

```python
y_pred = mlpc.predict(X_test_scaled)
accuracy_score(y_test, y_pred)
```

```python
cross_val_score(mlpc, X_test, y_test, cv = 10).mean()
```

```python
mlpc_params = {"alpha": [0.1, 0.01, 0.02],              #72.26 dakika sürdü.
              "solver" : ["lbfgs","adam","sgd"],
              "activation": ["relu","logistic"]}
mlpc = MLPClassifier()
mlpc_cv_model = GridSearchCV(mlpc, mlpc_params, 
                         cv = 10, 
                         verbose = 2)

mlpc_cv_model.fit(X_train_scaled, y_train)
```

```python
print("En iyi parametreler: " + str(mlpc_cv_model.best_params_))
```

```python
mlpc_tuned = MLPClassifier(activation = "logistic", 
              alpha = 0.01,
              solver = "adam")
mlpc_tuned.fit(X_train_scaled, y_train)
```

```python
y_pred = mlpc_tuned.predict(X_test_scaled)
accuracy_score(y_test, y_pred)
```

```python
cross_val_score(mlpc_tuned, X_test, y_test, cv = 10).mean()
```
